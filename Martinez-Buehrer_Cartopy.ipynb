# Explaining the Basics of Cartopy

This next line of code imports Cartopy and saves it to the alias ccrs, as well as import cartopy feature, which uses a collection of points, lines and polygons for plotting. I will also install matplot lib to use as my base for plotting

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
%matplotlib inline
```

The next step will be to create a basic map. I will do that with the following code. ax is set as a variable to plot the axes in PlateCarree, which is a cylindrical projection.

```python
ax = plt.axes(projection=ccrs.PlateCarree()) #sets the projection
ax.coastlines() #adds the coastlines of the continents and islands

plt.show() #displays the plot
```

The next thing I want to do is focus on the United States. I can do that by setting my plot to be inline with a defined latitude and longitude box. Since my location is also in the Continental United States, I want to add in the State borders and Canadian Provinces. I can do so with the following code

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-130, -60, 20, 50]) #sets the figure to show just the United States and surronding countries
ax.coastlines()
ax.add_feature(cfeature.STATES) #adds state borders

plt.show()
```

Since it doesn't look good to have the United States and Canada have a straight border, we can use Lambert Conformal to make it look more realistic. I also want the figure to be larger, more focused on the United States, and easier to read. We can change that using a simple matplotlib function

```python
fig = plt.figure(figsize=(9,6)) #changes the size of the figure
ax = plt.axes(projection=ccrs.LambertConformal())
ax.set_extent([-120, -73, 23, 50]) #I changed the coordinates to make the US more centered
ax.coastlines()
ax.add_feature(cfeature.STATES)

plt.show()
```

To add on to this map, we can make it look more like a map we see in textbooks. Let's add some color, the great lakes, and some rivers. We can do this using another cartopy features (which we identified as cfeature) function, as well as NaturalEarthFeature

```python
from cartopy.feature import NaturalEarthFeature
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection=ccrs.LambertConformal())
ax.add_feature(cfeature.LAND) #adds the beige coloring to land
ax.add_feature(cfeature.LAKES, alpha = 0.5) #adds the light blue coloring to lakes
ax.add_feature(cfeature.RIVERS) #adds the light blue coloring to rivers
ax.add_feature(cfeature.OCEAN) #adds the blue coloring to oceans
ax.add_feature(cfeature.STATES)
ax.set_extent([-120, -73, 23, 50])
ax.coastlines()

states = cfeature.NaturalEarthFeature(category='cultural',
                                            name='admin_1_states_provinces_lines',
                                            scale='50m', facecolor='none') #category is the type of dataset, name is the name of the dataset, scale is the scale of the dataset, and facecolor fills in the polygons 
ax.add_feature(states, edgecolor='gray')

plt.show()
```
